```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import numpy as np
np.random.seed(555)
sns.set()
```

# 様々なAR(1)を発生させる

```python
# 平均0, 分散1の正規分布から発生させた乱数100個
arr = np.random.normal(0,1,100)
```

```python
# 平均0, 標準偏差1の正規分布から発生させた乱数100個
wn = np.random.normal(0,1,100)
# 定数c
c = 2
# φ１
fai_1 = 0.8
# 初期値y0=0
map1 = [0]

for i in range(1,len(wn)):
    map1.append(c+fai_1*map1[i-1]+wn[i])
```

```python
plt.plot(map1)
```

```python
#　自己相関
sm.graphics.tsa.plot_acf(map1, lags = 10)
```

# 様々なAR(2)を発生させる

```python
# 平均0, 標準偏差1の正規分布から発生させた乱数100個
np.random.seed(555)
wn = np.random.normal(0,1,100)
# 定数c
c = 2
# φ１
fai_1 = 0.8
# φ2
fai_2 = 0
# 初期値y0=0
map1 = [0]

for i in range(1,len(wn)):
    map1.append(c+fai_1*map1[i-1]+fai_2*map1[i-2]+wn[i])
```

```python
plt.plot(map1)
```

```python
#　自己相関
sm.graphics.tsa.plot_acf(map1, lags = 20)
```
